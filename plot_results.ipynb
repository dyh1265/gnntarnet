```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([16, 32, 64, 128])

y1 = np.array([3.27, 2.22, 1.37, 1.03])
y2 = np.array([4.35, 2.11, 1.69, 1.72])
y3 = np.array([12.25, 7.24, 7.64, 8.37])
y4 = np.array([41.66, 27.46, 26.51, 29.19])

y1_eye = np.array([4.39, 2.59, 1.12, 0.84])
y2_eye = np.array([16.82, 7.31, 1.81, 0.87])
y3_eye = np.array([85.03, 36.27, 12.65, 4.60])
y4_eye = np.array([305.48, 127.72, 56.75, 25.17])

y1_tarnet = np.array([3.38, 3.22, 3.59, 3.60])
y2_tarnet = np.array([16.25, 15.66, 17.64, 11.04])
y3_tarnet = np.array([63.9, 59.74, 67.80, 31.41])
y4_tarnet = np.array([256.7, 244.0, 274.0, 142.57])
# Adding random errors to the data


errors_y1 = np.array([0.30, 0.22, 0.12, 0.03])
errors_y2 = np.array([0.69, 0.29, 0.18, 0.17])
errors_y3 = np.array([2.94, 1.19, 1.19, 1.23])
errors_y4 = np.array([11.56, 6.59, 5.87, 4.92])

errors_y1_eye = np.array([0.41, 0.26, 0.09, 0.04])
errors_y2_eye = np.array([2.58, 1.27, 0.25, 0.11])
errors_y3_eye = np.array([15.57, 11.57, 3.96, 0.84])
errors_y4_eye = np.array([53.94, 29.54, 15.43, 7.03])

errors_y1_tarnet = np.array([0.24, 0.21, 0.23, 0.24])
errors_y2_tarnet = np.array([1.50, 1.42, 1.53, 1.58])
errors_y3_tarnet = np.array([9.11, 8.01, 9.46, 7.68])
errors_y4_tarnet = np.array([40.94, 38.68, 42.95, 41.42])
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot the first set of data with errors
axs[0, 0].errorbar(x, y1, yerr=errors_y1, capsize=5, label='GNN-TARnet(graph)')
axs[0, 0].errorbar(x, y1_eye, yerr=errors_y1_eye, capsize=5, label='GNN-TARnet(ident.)')
axs[0, 0].errorbar(x, y1_tarnet, yerr=errors_y1_tarnet, capsize=5, label='TARnet')

# axs[0, 0].set_yscale('log')
axs[0, 0].set_xticks(x)
axs[0, 0].set_title('1 layer')
axs[0, 0].set_xlabel('Num samples')
axs[0, 0].set_ylabel('Sqrt(PEHE)')

# Plot the second set of data with errors
axs[0, 1].errorbar(x, y2, yerr=errors_y2, capsize=5, label='GNN-TARnet(graph)')
axs[0, 1].errorbar(x, y2_eye, yerr=errors_y2_eye, capsize=5, label='GNN-TARnet(ident.)')
axs[0, 1].errorbar(x, y2_tarnet, yerr=errors_y2_tarnet, capsize=5, label='TARnet')

# axs[0, 1].set_yscale('log')
axs[0, 1].set_xticks(x)
axs[0, 1].set_title('2 layers')
axs[0, 1].set_xlabel('Num samples')
axs[0, 1].set_ylabel('Sqrt(PEHE)')

# Plot the third set of data with errors
axs[1, 0].errorbar(x, y3, yerr=errors_y3, capsize=5, label='GNN-TARnet(graph)')
axs[1, 0].errorbar(x, y3_eye, yerr=errors_y3_eye, capsize=5, label='GNN-TARnet(ident.)')
axs[1, 0].errorbar(x, y3_tarnet, yerr=errors_y3_tarnet, capsize=5, label='TARnet')


# axs[1, 0].set_yscale('log')
axs[1, 0].set_xticks(x)
axs[1, 0].set_title('3 layers')
axs[1, 0].set_xlabel('Num samples')
axs[1, 0].set_ylabel('Sqrt(PEHE)')

# Plot the fourth set of data with errors
axs[1, 1].errorbar(x, y4, yerr=errors_y4, capsize=5, label='GNN-TARnet(graph)')
axs[1, 1].errorbar(x, y4_eye, yerr=errors_y4_eye, capsize=5, label='GNN-TARnet(ident.)')
axs[1, 1].errorbar(x, y4_tarnet, yerr=errors_y4_tarnet, capsize=5, label='TARnet')



axs[1, 1].set_yscale('log')
axs[1, 1].set_xticks(x)
axs[1, 1].set_title('4 layers')
axs[1, 1].set_xlabel('Num samples')
axs[1, 1].set_ylabel('Sqrt(PEHE)')

# Add legend to each subplot
for ax in axs.flat:
    ax.legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
